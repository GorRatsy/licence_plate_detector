```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import easyocr
```

```python
plate_class = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml')
```

```python
imgg = cv2.imread('/home/gor/Downloads/1—Å.jpg')
gray = cv2.cvtColor(imgg,cv2.COLOR_BGR2GRAY)
```

```python
plate_coord = plate_class.detectMultiScale(gray, 1.1,4)
x = plate_coord[0][0]
y = plate_coord[0][1]
w = plate_coord[0][2]
h = plate_coord[0][3]
plate = gray[y:y+h,x:x+w]
plate = cv2.cvtColor(plate,cv2.COLOR_BGR2RGB)
```

```python
cv2.rectangle(imgg, (x,y), (x+w, y+h), (255,0,0), 2)
plt.imshow(plate)
plt.show()
```
